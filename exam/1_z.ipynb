```python
import sys

def linear_search(numbers, target):
    for num in numbers:
        if num == target:
            return True
    return False

def binary_search(numbers, target):
    low = 0
    high = len(numbers) - 1

    while low <= high:
        mid = (low + high) // 2
        if numbers[mid] == target:
            return True
        elif numbers[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return False

search_functions = {
    "linear": linear_search,
    "binary": binary_search
}

search_type = input("Введите тип поиска (linear/binary): ")
if search_type not in search_functions  or search_type == '':
    print('Данного типа не существует')
    sys.exit()
numbers = input("Введите числа через пробел: ").split()
target = input("Введите число, которое нужно найти: ")

numbers = list(map(int, numbers))
search_func = search_functions[search_type]
result = search_func(numbers, int(target))
print("Результат:", result)


# 3 5 1 -2 4 10 35 12
```

# Задание 3 а)

```python
import matplotlib.pyplot as plt
import timeit
import numpy as np

def remove_elements_list(lst, k):
    for _ in range(k):
        lst.pop()

def remove_elements_array(arr, k):
    arr = np.delete(arr, np.s_[-k:], axis=0)

def measure_memory(obj):
    return obj.__sizeof__()

def measure_time(func, *args):
    start_time = timeit.default_timer()
    func(*args)
    elapsed_time = timeit.default_timer() - start_time
    return elapsed_time

def plot_graphs(N_values, k_values):
    for N in N_values:
        memory_data_list = []
        time_data_list = []
        memory_data_array = []
        time_data_array = []

        for k in k_values:
            lst = list(range(N))
            arr = np.array(range(N))

            memory_list = measure_memory(lst.copy())
            memory_data_list.append(memory_list)

            time_list = measure_time(remove_elements_list, lst.copy(), k)
            time_data_list.append(time_list)

            memory_array = measure_memory(arr.copy())
            memory_data_array.append(memory_array)

            time_array = measure_time(remove_elements_array, arr.copy(), k)
            time_data_array.append(time_array)

        plt.figure()
        plt.plot(k_values, memory_data_list, label=f'List - N={N}')
        plt.plot(k_values, memory_data_array, label=f'Array - N={N}')
        plt.title(f'Затрачено памяти для N={N}')
        plt.xlabel('Кол-во удаляемых элементов (k)')
        plt.ylabel('Память (байты)')
        plt.legend()

        plt.figure()
        plt.plot(k_values, time_data_list, label=f'List - N={N}')
        plt.plot(k_values, time_data_array, label=f'Array - N={N}')
        plt.title(f'Затрачено времени для N={N}')
        plt.xlabel('Кол-во удаляемых элементов (k)')
        plt.ylabel('Время (секунды)')
        plt.legend()

    plt.show()

N_values = [1000, 10000, 100000]
k_values = [1, 10, 100]

plot_graphs(N_values, k_values)
```

```python
import matplotlib.pyplot as plt
import timeit
from collections import deque

class QueueUsingList:
    def __init__(self):
        self.queue = []

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        return self.queue.pop(0)

    def sum_elements(self):
        return sum(self.queue)

class QueueUsingDeque:
    def __init__(self):
        self.queue = deque()

    def enqueue(self, item):
        self.queue.append(item)

    def dequeue(self):
        return self.queue.popleft()

    def sum_elements(self):
        return sum(self.queue)

def measure_memory(obj):
    return obj.__sizeof__()

def measure_time(func, *args):
    start_time = timeit.default_timer()
    func(*args)
    elapsed_time = timeit.default_timer() - start_time
    return elapsed_time

def plot_comparison(N_values):
    memory_data_list = []
    time_data_list = []
    memory_data_deque = []
    time_data_deque = []

    for N in N_values:
        queue_list = QueueUsingList()
        queue_deque = QueueUsingDeque()

        for i in range(N):
            queue_list.enqueue(i)
            queue_deque.enqueue(i)

        memory_list = measure_memory(queue_list)
        memory_data_list.append(memory_list)

        time_list = measure_time(queue_list.sum_elements)
        time_data_list.append(time_list)

        memory_deque = measure_memory(queue_deque)
        memory_data_deque.append(memory_deque)

        time_deque = measure_time(queue_deque.sum_elements)
        time_data_deque.append(time_deque)

    plt.figure()
    plt.plot(N_values, memory_data_list, label='List Queue')
    plt.plot(N_values, memory_data_deque, label='Deque Queue')
    plt.title('Затрачено памяти на суммирование элементов в очереди')
    plt.xlabel('Размер очереди (N)')
    plt.ylabel('Память (байты)')
    plt.legend()

    plt.figure()
    plt.plot(N_values, time_data_list, label='List Queue')
    plt.plot(N_values, time_data_deque, label='Deque Queue')
    plt.title('Затрачено времени на суммирование элементов в очереди')
    plt.xlabel('Размер очереди (N)')
    plt.ylabel('Время (секунды)')
    plt.legend()

    plt.show()

N_values = [1000, 5000, 10000]
plot_comparison(N_values)
```

# Задание 2 а)

```python
# Производная n**2 + 3*n = math.log(2) * (2 ** n) - 15 * n ** 2
# Производная 2^n - 5n^3 + 1' = 2*n + 3

import matplotlib.pyplot as plt
import numpy as np
import math

def f1(n):
    return math.log(2) * (2 ** n) - 15 * n ** 2

def f2(n):
    return 2*n + 3

n_values = np.linspace(0, 10, 100)

f1_values = f1(n_values)
f2_values = f2(n_values)


plt.plot(n_values, f1_values, label='f1(n) = n^2 + 3n')
plt.plot(n_values, f2_values, label='f2(n) = 2^n - 5n^3 + 1')

plt.title('Графики функций f1 и f2')
plt.xlabel('Количество элементов')
plt.ylabel('Значение функции')
plt.legend()
plt.grid(True)
plt.show()

# f1(n) растет быстрее => она является g(n)
```

## Длина (len):

Количество операций: O(1) - операция получения длины выполняется за постоянное время.
## Добавление элемента в конец (append):

Количество операций: O(1) в среднем (амортизированное время).
## Подсчет количества элементов (count):

Количество операций: O(n) - линейное время, где n - количество элементов в списке.
## Удаление элемента по индексу (pop):

Количество операций: O(n) - линейное время, так как при удалении элемента требуется перемещение оставшихся элементов.
## Получение элемента по индексу (indexing):

Количество операций: O(1) - операция получения элемента по индексу выполняется за постоянное время.
## Реверсия (reverse):

Количество операций: O(n) - линейное время, где n - количество элементов в списке.
## Проверка вхождения элемента (in):

Количество операций: O(n) - линейное время, так как может потребоваться проверка каждого элемента.
## Расширение массива путем добавления другого массива (extend):

Количество операций: O(k), где k - количество элементов в добавляемом массиве.
## Сравнение (==):

Количество операций: O(n) - линейное время, где n - количество элементов в списке.
## Вставка/добавление элемента в указанную позицию (insert):

Количество операций: O(n) в среднем (амортизированное время), так как может потребоваться перемещение оставшихся элементов.
## Умножение на число (*):

Количество операций: O(n) - линейное время, где n - количество элементов в списке.
## Изменение значения элемента (indexing):

Количество операций: O(1) - операция изменения значения элемента по индексу выполняется за постоянное время.
