```python
import matplotlib.pyplot as plt
import random

def generate_list(length, fraction):
    num_elements = int(length * fraction)
    return [random.randint(1, 100) for _ in range(num_elements)]

def plot_list_length_vs_fraction(lengths, fractions):
    for fraction in fractions:
        
        num_elements = [len(generate_list(length, fraction)) for length in lengths]
        plt.plot(lengths, num_elements, label=f'Доля чисел: {fraction}')
        print(num_elements)

    plt.xlabel('Размер листа')
    plt.ylabel('Элементы')
    plt.legend()
    plt.show()

lengths = list(range(10, 101, 10))
fractions = [0.1, 0.3, 0.5]

plot_list_length_vs_fraction(lengths, fractions)
```

```python
import time

arr = [1]*20

for i in range(len(arr)-1, -1, -1):
    start_time = time.time()
    arr.pop(i)
    end_time = time.time()
    elapsed_time = end_time - start_time
    print(f'Index {i}: Elapsed Time: {elapsed_time:.20f} seconds')
    print('Length:', len(arr), 'Size:', sys.getsizeof(arr))
```
