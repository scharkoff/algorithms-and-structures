```python
def fun(n):
    res = [0, 0, 1]
    for i in range(3, n):
        res.append((res[i-2] + res[i-1]))
    return res[-1]

print(fun(10))
        

```

```python
import time

count = 0

for i in range(10, 40, 2):
    for _ in range(3):
        t0 = time.perf_counter() 
        count += 1
        fun(i)
        print(i, ": ", time.perf_counter() - t0, end=" ")
        count += 1
    print()

print("Количество операций: ", count)
```
